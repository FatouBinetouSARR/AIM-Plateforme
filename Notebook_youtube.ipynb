```python
import re
import emoji
import nltk
from nltk.corpus import stopwords

# Télécharger les stopwords de NLTK si ce n'est pas déjà fait
nltk.download('stopwords')

# 1. Définir la fonction de nettoyage spécifique YouTube
def youtube_spam_specific_cleaner(text):
    """
    Nettoyage spécifique pour les commentaires YouTube
    """
    if not isinstance(text, str):
        return ""
    
    # Convertir en minuscules
    text = text.lower()
    
    # Supprimer les URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    
    # Supprimer les mentions @ (très commun dans les commentaires)
    text = re.sub(r'@\w+', '', text)
    
    # Garder les émojis (peuvent être informatifs) mais les convertir en texte
    text = emoji.demojize(text)
    
    # Supprimer les caractères répétés (ex: "cooooooool" -> "cool")
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # Supprimer la ponctuation mais garder certains caractères pour les émoticônes
    text = re.sub(r'[^\w\s:)(;/\-]', '', text)
    
    # Nettoyer les espaces multiples
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# 2. Définir la liste de stop words spécifiques au spam YouTube
youtube_stopwords = set(stopwords.words('english')).union({
    'watch', 'video', 'youtube', 'channel', 'subscribe',
    'like', 'comment', 'share', 'thanks', 'thank', 'please',
    'click', 'link', 'check', 'www', 'com', 'http', 'https',
    'hey', 'hi', 'hello', 'ok', 'yes', 'no', 'good', 'bad',
    'nice', 'great', 'awesome', 'love', 'lol', 'haha'
})

# 3. Définir la fonction pour supprimer les mots spécifiques YouTube
def remove_youtube_specific_words(text):
    if not isinstance(text, str):
        return ""
    
    words = text.split()
    words = [w for w in words if w not in youtube_stopwords]
    return ' '.join(words)

# 4. Maintenant, renommer les colonnes et appliquer le nettoyage
df_ytube.rename(columns={'CONTENT': 'text', 'CLASS': 'label'}, inplace=True)

# 5. Appliquer le nettoyage
print("Nettoyage en cours...")
df_ytube['cleaned_text'] = df_ytube['text'].apply(youtube_spam_specific_cleaner)
df_ytube['cleaned_text'] = df_ytube['cleaned_text'].apply(remove_youtube_specific_words)

# 6. Vérification
print("\n=== VÉRIFICATION APRÈS NETTOYAGE ===")
print(f"Nombre d'échantillons : {len(df_ytube)}")
print(f"Distribution des classes :")
print(df_ytube['label'].value_counts())
print(f"\nExemple avant/après nettoyage :")
print("Avant (150 premiers caractères) :")
print(df_ytube['text'].iloc[0][:150] if len(df_ytube) > 0 else "DataFrame vide")
print("\nAprès nettoyage :")
print(df_ytube['cleaned_text'].iloc[0][:150] if len(df_ytube) > 0 else "DataFrame vide")
```
